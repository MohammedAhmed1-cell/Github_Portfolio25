# **Python, Input and Type Casting Task**

```python
# Ask the user for the prices of two items
item1 = input("Enter the price of item 1: ")
item2 = input("Enter the price of item 2: ")

# The input() function returns strings, so we need to convert them to numbers
# We'll use float() since prices can include decimals
price1 = float(item1)
price2 = float(item2)

# Calculate the total
total = price1 + price2

# Display the total bill
print("Total:", total)

#print(f"Total: {total:.2f}")
print(f"The total bill price is: £ {total}0")

my_Name = "Mohammed"
print(f"My name is {my_Name}")
```
