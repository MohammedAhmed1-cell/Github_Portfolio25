# **Comparison operators**

```python
# Check equality (==) - Is the value on the left EQUAL to he value of the right
num1 = 10
print(num1 == 10)
print(num1 == 20)

# Check for inequality (!=) - i sthe value on the left NOT EQUAL to the value of the right
print(num1 != 10)
print(num1 != 20)

# Examples- check if value is greater than (>) - is the value on the left GREATER THAN the value on the right
age = 18
print(age > 17)

# Check if a value is less than (<) - is the value on the left LESS THAN the value on the right
print(age < 18)

# GREATER or LESS than equal to (>= / <=) is the value GREATER or LESS with the comparison valiue included
print( age >= 18)
print(age <= 18)
```

```python
# If Else statement - checks a condition and runs the associated if block if the condition is TRUE
# if it is FALSE it will run the Else block code

print("<<<<<< Before the if statment >>>>>>>")

age = 17

if age >= 18:
    print(f"Greate you're {age}, Go on in. Enjoy the film")
else:
  print(f"Sorry you're only {age}, you're not old enough to see this film")


print("<<<<<<< After the if statement >>>>>>>")
```

# **While loops**

```python
num = 0
loop_counter = 1

while num <= 50:
  print(f"Loop Number: {loop_counter}")
  loop_counter += 1
  print(num)
  num += 5
```

```python
# While Loop with break
i = 1
while i < 6:
  print(i)
  if i == 3:
    break
  i += 1

print()

# While Loop with continue
x = 0
while x < 6:
  x += 1
  if x == 3:
    continue
  print(x)
```

Let's play a â€œguess the number gameâ€ using while loop.


Write a program that:

Takes the number from user

Compares the number the user entered with a randomly generated number

Provide a message â€œLittle lowerâ€ or â€œLittle higherâ€ till the user guesses it correctly

```python
import random

# Generate a random number between 1 and 100
secret_number = random.randint(1, 100)

# Ask the user to guess the number
guess = int(input("Guess the number (between 1 and 100): "))

# Loop until the user guesses correctly
while guess != secret_number:
    if guess < secret_number:
        print("Little higher!")
    else:
        print("Little lower!")
    guess = int(input("Try again: "))

print("ðŸŽ‰ Congratulations! You guessed it right!")
```

# **Range function**

```python
# Range with 1 argument will start at zero and count as many times as you define
numbers = range(5)

# Range with 2 arguments - gives us a start and stop point for our range - stop is not included
numbers = range(10, 20)

# Range with 3 arguments - start, stop and step - Define the range and how much we want to increase by
numbers = range(0, 51, 5)
for i in numbers:
  print(i)
```

# **Lists**

```python
colours = ["red", "green", "blue", "yellow", "white", "black"] # red = 0

#Print entire list
print(colours)

#Print a specific item via its index - Index starts at 0
print(colours[0])

# Negative indexing  - starts from the right at -1 and counts backswards so -2, -3....
print(colours[-1]) # gold = -1 (-1 will be the last item)
print(colours[-5])

# Access a slice of a list - set a starting index and a non-inclusive end point, create new list with our slice
print(colours[2:5])

# Update the value of a list item - reference the list name and the idex of the item you want to update
# same as a variable we use assignment operator and set a new value
colours[0] = "silver"
print(colours)

# Add new list item to the end
colours.append("red")
print(colours)

# Loop over the list
for i in colours:
  print(i)
```
