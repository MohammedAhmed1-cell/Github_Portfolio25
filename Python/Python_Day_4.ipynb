# Convert Date to DateTime Format

```python
df['IMF_Year'] = pd.to_datetime(df['IMF_Year'], format='%Y', errors='coerce')
```

# Remove Whitespace In Text Data (Trim / Strip)

```python
df['Country/Territory'] = df['Country/Territory'].str.strip()
```

# Convert Data Types

```python
  # - tidy or transform columns
df['WorldBank_Estimate'] = pd.to_numeric(df['WorldBank_Estimate'], errors='coerce').fillna(0).astype(int)  # change the data type (e.g., text to number)
```

# Import MatPlotLib and Seaborn


```python
import matplotlib.pyplot as plt
import seaborn as sns
```

# Select only the numerical features for correlation analysis

```python
numerical_features = df.select_dtypes(include=['number'])
```

# Calculate a Correlation Matrix

```python
corr = numerical_features.corr()
```

# Plot a Heatmap

```python
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='GnBu')
plt.show()
```

# Plot a Barplot

```python
fig = plt.figure(figsize=(8, 5))
sns.barplot(x="IMF_Estimate", y="UN_Region", data=df, ci=None)
plt.show()
```

# Plot a Scattter Chart

```python
df.plot(x='UN_Region', y='UN_Estimate', kind='scatter', figsize=(10, 6), title="scatter")
```

# Plot a Bar Chat

```python
ax = sns.countplot(x="UN_Region", data=df[-(df.UN_Region=="World")])
ax.bar_label(ax.containers[0])

plt.title("  # of Countries")
plt.show()
```
